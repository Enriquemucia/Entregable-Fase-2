```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1. Dataset - Bank Marketing

```python
data = pd.read_csv("bank-additional-full.csv", sep=';')
data.head()
```

### 2.	Análisis exploratorio

```python
print(data.describe())
```

```python
data.isnull().mean()
```

```python
# Graficar la distribución de las variables numéricas
numerical_cols = data.select_dtypes(include=['int64', 'float64']).columns
data[numerical_cols].hist(bins=15, figsize=(20, 15), layout=(4, 3))
plt.suptitle('Distribución de las variables numéricas')
plt.show()
```

```python
# Graficar variables categóricas
categorical_cols = data.select_dtypes(include=['object']).columns
for col in categorical_cols:
    plt.figure(figsize=(5, 2))
    sns.countplot(data=data, x=col)
    plt.title(f'Distribución de {col}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Correlación entre variables numéricas
plt.figure(figsize=(12, 8))
correlation_matrix = data.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap de la correlación entre variables numéricas')
plt.show()
```
